# **Exploratory  Data Analysis and Visualization**

## **Q.1 What is the percentage of booking for each year?**

```python
# Calculate total number of bookings acorss different years 
years_booking = df['arrival_date_year'].value_counts().sort_index()
print(years_booking)
```

```python
# Percentage distribution of total booking across different years
df['arrival_date_year'].value_counts(normalize=True) * 100
```

```python
# Plot the graph for comparing year with total bookings done
plt.figure(figsize = (12,6), dpi=110)
sns.countplot(x='arrival_date_year', data=df, palette='colorblind')
plt.title('Yearly Comparison')
plt.xlabel('Year')
plt.ylabel('Total Bookings')
```

This plot clearly represents that bookings in year 2016 was highest 
compared to other years which is near about more than 56500. In year 2017 bookings fell down as compared to last year i.e. 40600 but this value was still quite better than it was in 2015 only about 22000.





```python
# Plot the graph by grouping by on hotel type for comparing year with total bookings done
plt.figure(figsize = (12,6), dpi=110)
sns.countplot(x='arrival_date_year', hue='hotel', data=df, palette='Paired')
plt.title('Year-wise and Hotel-wise Comparison')
plt.legend(title='Hotel Type')
plt.xlabel('Year')
plt.ylabel('Count')
```

```python
# Enlarging the pie chart
plt.rcParams['figure.figsize'] = 8,8

# Indexing labels. tolist() will convert the index to list for easy manipulation
labels = df['hotel'].value_counts().index.tolist()

# Convert value counts to list
sizes = df['hotel'].value_counts().tolist()

# As the name suggest, explode will determine how much each section is separated from each other 
explode = (0, 0.1)

# Determine colour of pie chart
colors = ['green','red']

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',startangle=90, textprops={'fontsize': 14})
```

Here, We can analyse that city hotels always have higher demand as compared to resort hotels.

# **Q.2 When people stay in hotels, how long do they usually stay?**

```python
# Let us find out night stay duration of customers
total_nights_stays = (df['stays_in_weekend_nights']+ df['stays_in_week_nights']).value_counts().head(10).sort_index()
print(total_nights_stays)
```

```python
# Plot the graph of number of nights people wants to stay against total number of bookings
total_nights_stays.plot.bar(color='blue')
plt.rcParams['figure.figsize'] = (13,7)
plt.grid()
plt.title('Night Stay Duration')
plt.ylabel('Total number of bookings')
plt.xlabel('Number of nights')
```

*   Generally, customers prefer to stay one, two, three, or four nights; however, bookings for two and three nights have a higher demand.

*   In order to get the best daily rate, we should consider staying for two to four days, since more people prefer the same length of stay. Because long-term stay offers will consume their rooms and new customers will not be able to take advantage of the rooms, hotels are also offering mid-term accommodation to generate extra profits.




```python
# Observing peoples interest to stay according to type of hotel with graph
df.loc[:,'total_nights'] = df['stays_in_weekend_nights']+ df['stays_in_week_nights']

fig, ax = plt.subplots(figsize=(12,6))
sns.countplot(x='total_nights', hue='hotel', data=df, order = df.total_nights.value_counts().iloc[:10].index)
ax.set_title('Hotel wise night stay duration')
ax.set_xlabel('No of Nights')
ax.set_ylabel('Count')
```


*   According to this visual representation, people are interested in city hotel to stay for two, three, one and four days respectively whereas for resort hotel people are interested to stay one, seven, two and three days respectively.
